```python
import pandas as pd
from os import path, listdir
from pathlib import Path
import proc_model
import matplotlib.pyplot as plt
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

#Min_distance

```python
data_dir = path.join(path.dirname(proc_model.__file__), 'outputs')
home = str(Path.home())
output_dir = path.join(home, 'Documents', 'bachelor', 'article', 'pm_results')
input_files = listdir(data_dir)
SAVE = True
```

```python
#read dataframe
filename = [f for f in input_files if 'mindist' in f][0]
df = pd.read_csv(path.join(data_dir, filename))
df['mindist'] = df['city'].apply(lambda x:x[-2:])
# print(df.head(5))
```

###Degree

```python
fig, axes = plt.subplots(ncols=2,  sharex=True, figsize=(28, 12))

ax = axes[0]
df.plot(kind='line', x='mindist', y='num_nodes', label='nodes', ax=ax)
df.plot(kind='line', x='mindist', y='num_edges', label='edges', ax=ax)
ax.set_xlabel('Minimum distance, [m]', labelpad=10)
ax.set_ylabel('Counts, [n] ', labelpad=10)
ax.set_yscale('log')
ax.legend(loc=2)
ax.set_title('Number of nodes and edges')
ax = axes[1]
df.plot(kind='line', x='mindist', y='node_density_km', label='nodes', ax=ax)
df.plot(kind='line', x='mindist', y='edge_density_km', label='edges', ax=ax)
ax.set_xlabel('Minimum distance, [m]', labelpad=10)
ax.set_ylabel('Density, [$n/km^2$]', labelpad=10)
ax.set_yscale('log')
ax.set_title('Node and edge density')

if SAVE:
    plt.savefig(path.join(output_dir, 'mindist1min.eps'), bbox_inches='tight')
```

###Degree

```python
fig, axes = plt.subplots(ncols=2,  sharex=True, figsize=(28, 12))

ax = axes[0]
df.plot(kind='line', x='mindist', y='degree_avg', label='node degree', ax=ax)
df.plot(kind='line', x='mindist', y='area_km', label='map area', ax=ax)
ax.set_xlabel('Minimum distance, [m]', labelpad=10)

ax.set_title('Node degree and map area.')
ax = axes[1]
df.plot(kind='line', x='mindist', y='edge_length_avg', label='average edge length', ax=ax)

ax.set_title('Average edge length')
ax.set_xlabel('Minimum distance, [m]', labelpad=10)
if SAVE:
    plt.savefig(path.join(output_dir, 'mindist2.eps'), bbox_inches='tight')
```
