```python
import pandas as pd
from os import path, listdir
from pathlib import Path
import proc_model
import matplotlib.pyplot as plt
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
data_dir = path.join(path.dirname(proc_model.__file__), 'outputs')
home = str(Path.home())
output_dir = path.join(home, 'Documents', 'bachelor', 'article', 'pm_results')

# print(df.head(5))
SAVE = True
```

```python
#read dataframe
filename = path.join(data_dir, 'scc.csv')
df = pd.read_csv(path.join(data_dir, filename))
df['size_m'] = df['city'].apply(lambda x: int(x[-4:]))

fig, axes = plt.subplots(ncols=2,  sharex=True, figsize=(28, 12))
ax = axes[0]
df.plot(kind='line', x='size_m', y='num_nodes', label='nodes', ax=ax)
df.plot(kind='line', x='size_m', y='num_edges', label='edges', ax=ax)
ax.set_xlabel('Min distance, [m]', labelpad=10)
ax.set_title('Number of nodes and edges')

ax = axes[1]
df.plot(kind='line', x='size_m', y='area_km', label='area, [$km^2$]', ax=ax)
df.plot(kind='line', x='size_m', y='node_density_km', label='node density, [$n/km^2$]', ax=ax)
ax.set_xlabel('Min distance, [m]', labelpad=10)
ax.set_title('Node and edge density ')
if SAVE:
    plt.savefig(path.join(output_dir, 'border_xy.eps'), bbox_inches='tight')
```
